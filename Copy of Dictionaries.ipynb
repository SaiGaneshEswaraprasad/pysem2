# Introduction

A dictionary is a data structure in which we store values as a pair of key and value.

Each item's key and its value are separated by a colon ( : ).

Each item is separated by a comma.

All the items of the dictionary are enclosed within curly brackets {}.

Syntax:
```
dictionary_name = {key1:value1, key2:value2, key3: value3}
```

Alternative way:

````python
''''
dictionary_name = {key1:value1, key2:value2, key3: value3}

If there are many keys and values in the dictionaries, then it can be written
as just one key-value pair on a line to make the code easier to read and
understand.

dictionary_name = {key1:value1, 
                   key2:value2, 
                   key3: value3,
                   }
'''
````

***Keys in the dictionary must be unique and be of any immutable data type.***

***Dictionaries are not sequences rather they are mappings.***

# Deleting items

We can delete one or more items in a dictionary using the keyword ***del***

Syntax:
```
del dictionary_var[key]
```

There are three more ways to delete the key-value pair
* clear()  -> deletes all the key-values of the dictionary
* pop() -> deletes the particuar key-value pair
* popitems() -> deletes randomly any key-value pair

````python
student = {'Rollno':1, 
           'Name':'Raj', 
           'Branch':'CSE', 
           'CGPA':8.9}
print(student)
del student['Rollno']
print(student)

student.pop('CGPA')
print(student)

student.clear() # -> deletes all the key-value pairs in 
                #dictionary student
print(student)
````

# Dictionary comprehensions 

it is another way of creating a dictionary.

It is syntatic construct which creates a dictionary based on existing dictionary.

Syntax:
```
D = {expression for variable in sequence [if condition]}
```

````python
d = {x: 2*x for x in range(1,10) if x%2 != 0}
print(d)

d = {x:x%2 for x in range(1,10)}
print(d)
d = {x:x**2 for x in range(1,10)}
print(d)
````

Key points:

# Sorting Items in a Dictionary

* keys() returns a list of all the keys used in the dictionary in an arbitrary order.

* sorted() function sorts the keys.

````python
student = {'Rollno':1, 'Name':'Raj', 'Branch':'CSE', 'CGPA':8.9}
student['Rank'] = 2
student['Branch'] = 'CSE-CS'
print(student)
print(student.keys()) #-> prints the keys in the dictionary
print(sorted(student.keys())) #-> sorts the keys
````

# Nested Dictionary

We can define a dictionary inside another dictionary

````python
class_set_B = {'Shiv':{'Python':90, 'Physics':89,'Maths':93},
               'Sadhvi':{'Python':87, 'Physics':85,'Maths':95},
               'Krish':{'Python':70, 'Physics':95,'Maths':78}
               } 

for k, v in class_set_B.items():
  print(k,':',v)

print(len(class_set_B))
````

# Difference bewtween a List and a Dictionary

* A List is an ordered set of items.A Dictionary is a set of matching of one item(key) with another(value).
* In a List, indexing is used to access a particluar item, where these indices are the numbers.In a Dictionary, we can use any immutable of value as an index.
* Lists are used to look up a value, whereas a Dictionary is used to take one value and look up another value. Dictionary is also know as ***lookup table***.
