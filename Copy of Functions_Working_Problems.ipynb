```python
'''
13.Write a menu driven program using function to perform calculator operations
such as adding, substracting, multiply and dividing two integers.
'''
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        print("Error: Division by zero")
        return None
    else:
        return x / y

def calculator():
    print("Menu:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("0. Exit")

    while True:
        choice = int(input("Enter your choice: "))
        if choice == 0:
            break
        elif choice < 1 or choice > 4:
            print("Invalid choice")
        else:
            x = int(input("Enter the first integer: "))
            y = int(input("Enter the second integer: "))
            if choice == 1:
                result = add(x, y)
                print("{0} + {1} = {2}".format(x, y, result))
            elif choice == 2:
                result = subtract(x, y)
                print("{0} - {1} = {2}".format(x, y, result))
            elif choice == 3:
                result = multiply(x, y)
                print("{0} * {1} = {2}".format(x, y, result))
            elif choice == 4:
                result = divide(x, y)
                print("{0} / {1} = {2}".format(x, y, result))

calculator()
```

```python
'''
18.Write a program using the function (C(n,r)) to calculate the compound
interest for the given principal, rate of interest, and number of years.
'''
import math

def C(n, r):
    if n < r:
        return None  # n must be greater than or equal to r
    else:
        return math.factorial(n) // (math.factorial(r) * math.factorial(n-r))

def compound_interest(P, r, n, t):
    A = P * ((1 + r/n) ** (n*t))
    return A

# Take user input for P, r, n, and t
P = float(input("Enter the principal amount: "))
r = float(input("Enter the annual interest rate as a decimal: "))
n = int(input("Enter the number of times interest is compounded per year: "))
t = float(input("Enter the number of years: "))

# Calculate the compound interest
C_n_r = C(n, 3)  # choose 3 from n
interest = compound_interest(P, r, n, t)

# Print the result
if C_n_r is None:
    print("Error: n must be greater than or equal to r")
else:
    print(f"Compound interest = {C_n_r * interest:.2f}")
```

```python
'''
19. Write a program using a function that returns the surface area and 
volume of a sphere.
'''
import math

def sphere_properties(radius):
    volume = (4/3) * math.pi * (radius ** 3)
    surface_area = 4 * math.pi * (radius ** 2)
    return volume, surface_area

# Take user input for the radius
radius = float(input("Enter the radius of the sphere: "))

# Calculate the sphere properties
volume, surface_area = sphere_properties(radius)

# Print the results
print(f"Volume of the sphere = {volume:.2f}")
print(f"Surface area of the sphere = {surface_area:.2f}")
```

```python
'''
21. Write a program to calculate exp(x,y) using recursion and without recursion.
'''
def exp_iterative(x, y):
    result = 1
    for i in range(abs(y)):
        result *= x
    if y < 0:
        result = 1 / result
    return result
x = int(input('enter a number:'))
y = int(input('enter a number:'))
result = exp_iterative(x, y)
print(f"{x} raised to the power {y} is {result:.2f}")
```

```python
'''
22. Write a program to print the Fibonacci series using recursion and without recurison.
'''
def fib(x):
  if x==1:
    return 0
  elif x==2:
    return 1
  else:
    return fib(x-1)+fib(x-2)
x=int(input('x: '))
fib(x)
```

```python
'''
23. Write a function that converts temperature given in Celsius to Faherenheit
an vice-versa.
'''
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius
temp = float(input("Enter the temperature: "))
unit = input("Enter the unit of measurement (C/F): ")
if unit == "C":
    converted_temp = celsius_to_fahrenheit(temp)
    print(f"{temp:.1f}°C is equal to {converted_temp:.1f}°F")
elif unit == "F":
    converted_temp = fahrenheit_to_celsius(temp)
    print(f"{temp:.1f}°F is equal to {converted_temp:.1f}°C")
else:
    print("Invalid unit of measurement. Please enter 'C' or 'F'.")
    
```
