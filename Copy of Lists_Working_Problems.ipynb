```python
'''
Write a program that passes a list to a function that scales each element in the list
by a factor of 10. Print the list values at different stages to show that change made
to one list is automatically refelected in the other list.
'''
def scale_list(list_to_scale):
    for i in range(len(list_to_scale)):
        list_to_scale[i] *= 10
        print(list_to_scale)

my_list = [1, 2, 3, 4, 5]

print("Original list:", my_list)
scale_list(my_list)
print("Scaled list:", my_list)
```

```python
'''
Write a program that has a list of both positive and negative numbers. Create another
list using filter() that has only positive values.
'''
original_list = [-5, 10, -15, 20, -25, 30]
positive_numbers = list(filter(lambda x: x > 0, original_list))

print("Original List:", original_list)
print("Positive Numbers:", positive_numbers)
```

```python
'''
Write a program to find the largest value in a list using reduce()function.
'''
import functools
def gr(x,y):
  if x>y:
    return x
  else:
    return y
num_list = [1,2,3,4,5]
print(functools.reduce(gr, num_list))
```
