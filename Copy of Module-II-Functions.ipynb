# Key points to remember

You can have a variable with the same name as that of a global variable in the program.
In such a case a new local variable of that name is created which is different from the gloable variable.

````python
var = 'good'
def show():
  var = 'Morning'
  print('In function var is - ',var)

show()
print('Outside function, var is -',var)
````

If we have global variable and then create another global variable using global statement, then changes made in the variable will be refelected everywhere in the program.

````python
var = 'good'
def show():
  global var
  var = 'Morning'
  print('In function var is - ',var)

show()
print('Outside function, var is -',var)
var = 'Fantastic'
print('Outside function, after modification var is -',var)
````

In case of nested functions, the inner function can access variables defined in both outer as well as inner function, but the outer function can access variables defined only in the outer function.

````python
def outer_func():
  outer_var = 10
  def inner_func():
    inner_var = 20
    print('Outer variable', outer_var)
    print('Inner variable', inner_var)
  
  inner_func()
  print('Outer variable = ', outer_var)
  print('Inner Variable = ', inner_var)


outer_func()
````

If a variable in the inner function is defined with the same name as that of a variable defined in the outer function, then a new variable is created in the inner function.

````python
def outer_func():
  var = 10
  def inner_func():
    var = 20
    print('Inner variable = ', var)
  
  inner_func()
  print('Outer variable = ', var)


outer_func()
````

# THE return Statement

A return statement is used to return some value of the caller.

Syntax:
```
return [expression]
```
expression is optional.

If expression is present, it is evaluated and the resultant value is returned to the calling function.

If no expression present, it will return ***None***.

## Having a return statement is not mandatory.

The return statement is used for two things:


*   Return a value to the caller
*   To end and exit a function and go back to its caller

## Points to remember
*  The return statement must appear within the function
*  Once a value is returned from a function, it immediately exits that function. So, any code retunred after the return statement is never executed.






````python
def cube(x):
  return x**3

num = int(input('Enter the number'))
result = cube(num)
print('Cube of ',num,' is ',result)
````

# LAMBDA FUNCTIONS OR ANONYMOUS FUNCTIONS

These functions are not declared as other functions using the keyword ***def*** .

They are created using the keyword ***lambda***. 

***lambda*** are throw-away functions, i.e. they are just the needed where they have been created and can be used anywhere a function is needed.


# Key point to remember

Lambda functions contains only a single line. Syntax is as shown below.
```
lambda arguments : expression
```
The arguments contain a comma separated list of arguments and the expression is an arithmetic expression that uses the arguments.



````python
sum = lambda x, y: x+y
#print('Sum = ',sum(5,4))

num1 = int(input('Enter the 1st number'))
num2 = int(input('Enter the 2nd number'))

print('Sum = ',sum(num1,num2))
````

# Key points to remember
Lambda functions
* have no name
* can take any number of arguments
* can return just one value in the form of an expression
* definition dose not have an explict ***return*** statement but it always contains an expression which is returned.
* One-line version of a function and hence cannot contain multiple expressions.
* They cannot access variables other than those in their parameter list.
* cannot even access global variable.
* Lambda functions can be passed as arguments in other functions. 
* Lambda functions are used along with built-in functions like filter(), map(), reduce().
* Lambda functions can be used with regular functions.

````python

# Lambda function assigned to a variable
twice = lambda x: 2*x

print(twice(9))


# lambda function not assigned to any variable

print('7 power of 4 is ',(lambda x: x**4)(7))
````

# RECURSIVE FUNCTIONS
A function defined as a function that calls itself to solve a smaller version of its task until a final call is made which does not requrie a call to itself.

A recursvie solution has two major cases:


*   ***base case*** : in which the problem is simple enough to be solved directly without making any further calls to the same function.
*   ***recursive case***: in which (i) the problem at hand is divided into simpler sub-parts. (ii) the function calls itself but with sub-part of the problem obtained in the firt step. (iii) the result is obtained by combining the solution of simple sub-parts.



````python

def small(a, b):
  if(a<b):
    return a
  else:
    return b


sum = lambda x, y : x+y
diff = lambda x, y: x-y

num1 = int(input('Eneter the first number'))
num2 = int(input('Eneter the second number'))
num3 = int(input('Eneter the third number'))
num4 = int(input('Eneter the fourth number'))

print('Smaller of the two numbers is ', small(sum(num1,num2), diff(num3, num4)))
````

````python
# passing lambda function as an argument to normal functions

def func(f,n):
  print(f(n))


twice = lambda x: x*2
thrice = lambda x: x*3

func(twice,4)
func(thrice, 10)
````
