# List Comprehensions
Python supports computed lists called list comprehensions.
Syntax:
```
List = [ expression for variable in sequence]
```
Here the expression is evaluated once, for every item in the sequence.

List comprehensions help to create lists in a concise way.

It is useful when each element is obtained by applying some operations to each member of another sequence or iterable.

List comprehension is also used to create a subsequence of those elements that satisfy a certain condition.



````python
# normal way to create a list of cubes
cubes = []

for i in range(11):
  cubes.append(i**3)
print('Cubes of numbers from 1-10',cubes)
````

````python
# Finding the cubes using List comprehenshions

cubes1 = [i**3 for i in range(11)]
print(cubes1)
````

````python
'''Using list comprehenshions to combine two elements
In other words, creating a list of tuples.
'''
numbers = [1,2,3,4,5]
#alphabets = [1,2,3,4,5]
alphabets = ['a','b','c','d','e']
print([(x,y) for x in numbers for y in alphabets])
````

````python
'''Using list comprehenshions to combine two elements
only if they are not same
'''
numbers = [1,2,3,4,5]
alphabets = [1,2,3,4,5]

print([(x,y) for x in numbers for y in alphabets if x!=y])
````

Functional programming decomposes a problem into set of functions.

* map()

#filter()
It constructs a list from those elements of the list for which a function returns True.
```
filter(function, sequence)
```
filter() function returns a sequence that contains items from the sequence for which the function is True.
It sequence is string , Unicode or a tuple, then the result will be of the same type, otherwise always a list

````python
# list of number divisible by 2 or 4 using list comperhension
def check(x):
  if (x%2 == 0 or x %4 ==0):
    return 1

evens = list(filter(check, range(2,22)))
print(evens)
````

````python
# prime numbers
def prime(num):
  factors = 0
  for i in range(1,num):
    if num % i ==0:
      factors+=1
  if factors == 1:
    return 1

primes = list(filter(prime, range(1,20)))
print(primes)
````

````python
# finding perfect numbers
def perfect(num):
  sum = 0
  for i in range(1,num):
    if num % i ==0:
      sum = sum+i
  if sum == num:
    return 1

perfect_nums = list(filter(perfect,range(1,100)))
print(perfect_nums)
````

#map() function
map() function applies a particular function to every element of a list. 
```
map(function, sequence)
```
It returns the modified list.
The map() function calls function(item) for each item in the sequence and returns a list of the return values.

````python
# doubling the each item of the list.

def mul_2(x):
  return x * 2

num_list = [10,20,30,40,50]
print('original list', num_list)
update_list = list(map(mul_2,num_list))
print('List after doubling each number in the list:')
print(update_list)
````

````python
# finding the cubes with in range
def power(num):
  return num**3


cubes = list(map(power, range(1,10)))
print(cubes)
````

More than one sequenc can also be passed in the map() function, and following points to keep in mind:
* The function must have as many arguments as there are sequences.
* Each argument is called with the corresponding item from each sequence (None if one sequence is shorter than another).

````python
def add(x,y):
  return x+y

list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
list3 = list(map(add,list1,list2))
print(list3)
````

# reduce() function
The reduce() function returns a single value generated by called the function on the first two items of the sequence, then on the result and the next time, so on.
```
reduce(function, sequence)
```

#Points to remember to use reduce() function

* If there is only one item in the sequence, then it is returned
* If the sequence is empty an exception is raised.
* Creating a list in a very extensive range will generate a MemoryError or OverflowError.


````python
# import the functools 
import functools
def add(x,y):
  return x + y


num_list = [1,2,3,4,5]
print('Sum of the values in the list:')
print(functools.reduce(add, num_list))
````
