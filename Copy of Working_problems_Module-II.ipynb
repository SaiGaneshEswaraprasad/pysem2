```python
'''
18. Python function that accepts a string and calculates the number of uppercase letters and lowercase letters.
'''
def countul(string):
    upper_count = 0
    lower_count = 0
    for char in string:
        if char.isupper():
            upper_count += 1
        elif char.islower():
            lower_count += 1
    return (upper_count, lower_count)

a=input("enter a string:")
counts = countul(a)
print(counts)
```

```python
'''
30. Write a program that prompts user to enter a string and
returns in alphabetical order, a letter and its frequency of
occurrence in the string.(Ignore case)
'''
string = input("Enter a string: ")
string = string.lower().replace(" ", "")
frequency = {}
for char in string:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
sorted_keys = sorted(frequency.keys())
for key in sorted_keys:
    print(f"{key}: {frequency[key]}")
```

```python
'''
31. Write a program has a dictionary of your friends name
(as keys) and thier birthdays. Print the items in the 
dictionary in a sorted order.Prompt the user to enter a
name and check if it is present in the dictionary.
If the name does not exist, then ask the user to enter DOB. 
Add the details in the dictionary.
'''
friends = {'Klaus': '1990-05-10', 'Elijah': '1992-07-15', 'Hope': '1988-12-03', 'Damon': '1995-02-28'}
print("List of friends and their birthdays (sorted by name):")
for friend, dob in sorted(friends.items()):
    print(friend, dob)
name = input("\nEnter a friend's name: ")
if name in friends:
    print(f"{name}'s birthday is on {friends[name]}")
else:
    dob = input(f"{name} is not in the dictionary. Enter {name}'s birthday (YYYY-MM-DD format): ")
    friends[name] = dob
    print(f"Added {name} ({dob}) to the dictionary.")
print("\nList of friends and their birthdays (sorted by name):")
for friend, dob in sorted(friends.items()):
    print(friend, dob)
```

```python
'''
32. Write a program that displays a menu and its price.
Take the order from the customer. Check if the ordered product
is in the menu. In case it is not there, the customer should
be asked to reorder and if it is present, then product
should be added in the bill.
'''
menu = {'coffee': 2.50, 'tea': 1.75, 'sandwich': 5.50, 'salad': 7.25, 'cake': 3.75}
print("Menu:")
for item, price in menu.items():
    print(f"{item.capitalize()}: ${price:.2f}")
bill = 0.0
while True:
    order = input("\nEnter an item from the menu or type 'done' to finish: ").lower()
    if order == 'done':
        break
    if order in menu:
        bill += menu[order]
        print(f"Added {order} (${menu[order]:.2f}) to the bill. Current bill: ${bill:.2f}")
    else:
        print(f"Sorry, we don't have {order}. Please choose an item from the menu.")
print(f"\nTotal bill: ${bill:.2f}")
```
