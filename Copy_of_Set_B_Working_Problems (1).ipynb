```python
'''
5. Write a program that counts the number of lowercase characters, uppercase characters and digits
in a string entered by the user.
'''
string = input("Enter a string: ")
lowercase_count = 0
uppercase_count = 0
digit_count = 0
for char in string:
    if char.islower():
        lowercase_count += 1
    elif char.isupper():
        uppercase_count += 1
    elif char.isdigit():
        digit_count += 1
print("Lowercase characters:", lowercase_count)
print("Uppercase characters:", uppercase_count)
print("Digits:", digit_count)
```

```python
'''
11. Write a program that prompts the user to enter numbers.
Once the user enters -1, it displays the count, sum and average of even numbers
and that of odd numbers.
'''
even_count = 0
even_sum = 0
odd_count = 0
odd_sum = 0
number = int(input("Enter a number (-1 to quit): "))

while number != -1:
    if number % 2 == 0:
        even_count += 1
        even_sum += number
    else:
        odd_count += 1
        odd_sum += number
    number = int(input("Enter a number (-1 to quit): "))

if even_count > 0:
    even_average = even_sum / even_count
    print(f"Even numbers count: {even_count}, sum: {even_sum}, average: {even_average}")
else:
    print("No even numbers entered.")

if odd_count > 0:
    odd_average = odd_sum / odd_count
    print(f"Odd numbers count: {odd_count}, sum: {odd_sum}, average: {odd_average}")
else:
    print("No odd numbers entered.")
```

```python
'''
24. Write a program to print whether  every number in a range is prime or composite.
'''
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

for num in range(start, end + 1):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                print(num, "is a composite number.")
                break
        else:
            print(num, "is a prime number.")
    else:
        print(num, "is not a prime number.")
```

```python
'''
27. Write a program to generates the following pattern.
Hint : no of rows - 5, no of cols - 5
$ * * * *
* $     *
*   $   *
*     $ *
* * * * $
'''
for row in range(5):
    for col in range(5):
        if (row == 0 or row == 4) or (col == 0 or col == 4):
            if row==col:
              print('$',end=" ")
            else:
              print("*", end=" ")
        else:
            if row==col:
              print('$',end=" ")
            else:
              print(" ", end=" ")
    print()
```

```python
'''
35. An employee's total weekly pay is calculated by multiplying the hourly wage and number
of regular hours plus any overtime pay which in turn is calcluated as total overtime
hours multiplied by 1.5 times the hourly wage.
Write a program that takes as inputs the hourly wage, total regular hours, and total 
overtime hours, and prints an employee's total weekly pay.
'''
hourly_wage = float(input("Enter the hourly wage: "))
regular_hours = float(input("Enter the number of regular hours: "))
overtime_hours = float(input("Enter the number of overtime hours: "))

total_pay = (regular_hours * hourly_wage) + (overtime_hours * (hourly_wage * 1.5))

print("Total weekly pay:", total_pay)
```
